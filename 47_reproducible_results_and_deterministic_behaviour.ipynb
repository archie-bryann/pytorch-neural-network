```python
import torch
import numpy as np
import random
```

```python
# make the behaviour deterministic, sacrifising for speed (for debugging, not training)
seed=42
torch.manual_seed(seed)
np.random.seed(seed)
random.seed(seed)
```

```python
# make the behaviour deterministic, sacrifising for speed (for debugging, not training)
# if using cuda
torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```
