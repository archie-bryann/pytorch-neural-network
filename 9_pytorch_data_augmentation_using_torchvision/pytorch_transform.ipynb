```python
# Imports
import torch
import torchvision.transforms as transforms # Transformations we can perform on our dataset
from torchvision.utils import save_image
from customDataset import CatsAndDogsDataset
```

```python
# Load Data
# https://docs.pytorch.org/vision/0.9/transforms.html

# my_transforms = transforms.ToTensor()
my_transforms = transforms.Compose([ # does it in the order below
    transforms.ToPILImage(),
    transforms.Resize((256, 256)),
    transforms.RandomCrop((224, 224)),
    transforms.ColorJitter(brightness=0.5),
    transforms.RandomRotation(degrees=45),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.05),
    transforms.RandomGrayscale(p=0.02),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.0, 0.0, 0.0], std=[1.0, 1.0, 1.0]) # (value - mean) / std # Note, this does nothing.
])

dataset = CatsAndDogsDataset(csv_file = 'custom_datasets/cats_dogs.csv', root_dir = 'custom_datasets/cats_dogs_resized',
                            transform = my_transforms)
```

```python
img_num = 0
for _ in range(10):
    for img, label in dataset:
        # print(img.shape)
        save_image(img, 'transformed/img' + str(img_num) + '.png')
        img_num += 1
```
